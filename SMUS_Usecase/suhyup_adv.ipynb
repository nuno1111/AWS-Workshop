```python
# suhyup_adv.csv를 읽어서 Pandas DataFrame으로 변환
import pandas as pd
df = pd.read_csv('suhyup_adv.csv')
df.head()  # DataFrame의 첫 5행을 출력
```

```python
from strands import Agent

def analyze_advertisement_reason(title, content):
    """
    광고성 문자의 사유를 분석하는 함수
    """
    
    prompt = f"""    
<Instruction>
- <title>과 <content>는 광고성 문자의 제목과 내용입니다. message_title,message_contents를 확인해서 정보성 문자가 아닌 광고성 문자인 사유를 작성해주세요.
- 정보성 메시지의 정의는 아래 <정보성 메시지 정의> 를 확인해주세요.
- 결과는 앞뒤 부가설명을 제외하고, 사유만 작성해주세요.
</Instruction>

<title>
{title}
</title>

<content>
{content}
</content>

<정보성 메시지 정의>
정보성 메시지는 아래와 같은 내용에 해당하는 행위에 대해 발송이 허용됩니다.

정보성 메시지에 해당하는 예시
금융상품 및 관련 업무에 대한 은행의 이용정보

예: 주소, 연락처, 홈페이지 등

유인 문구가 없는 단순 정보

특정 금융상품 이용고객에게 안내

가입 상품에 관한 현황(이자율 변동, 상품조건 변경 등) 안내

관련 법규의 제정 및 개정 등에 따른 제도 변경 안내

상담 과정에서 고객의 질문에 의한 정보성 메시지 제공

정보성 메시지에 해당하지 않는 예시
금융상품명 노출, 특정 상품 광고, 금융거래 유인, 금융상품 가입 권유 등 광고성 메시지는 원천 차단

즉, 정보성 메시지는 단순 정보 제공이나 제도 안내, 고객 문의에 대한 응답 등 광고·권유 성격이 없는 메시지를 의미합니다.
</정보성 메시지 정의>
"""
    
    agent = Agent()
    result = agent(prompt)
    
    return result.message['content'][0]['text'] 
    
def suggest_informational_improvement(title, content):
    """
    정보성 문자의 개선 방안을 제안하는 함수
    """
    
    prompt = f"""    
<Instruction>
- <title>과 <content>는 광고성 문자가 아닌 정보성 문자의 제목과 내용입니다.
- 광고성 표현이 있다면, 정보성 표현으로 수정해주세요.
- 정보성 메시지의 정의는 아래 <정보성 메시지 정의> 를 확인해주세요.
- 결과는 앞뒤 부가설명을 제외하고, 제목과 내용만 작성해주세요.
</Instruction>

<title>
{title}
</title>

<content>
{content}
</content>

<정보성 메시지 정의>
정보성 메시지는 아래와 같은 내용에 해당하는 행위에 대해 발송이 허용됩니다.

정보성 메시지에 해당하는 예시
금융상품 및 관련 업무에 대한 은행의 이용정보

예: 주소, 연락처, 홈페이지 등

유인 문구가 없는 단순 정보

특정 금융상품 이용고객에게 안내

가입 상품에 관한 현황(이자율 변동, 상품조건 변경 등) 안내

관련 법규의 제정 및 개정 등에 따른 제도 변경 안내

상담 과정에서 고객의 질문에 의한 정보성 메시지 제공

정보성 메시지에 해당하지 않는 예시
금융상품명 노출, 특정 상품 광고, 금융거래 유인, 금융상품 가입 권유 등 광고성 메시지는 원천 차단

즉, 정보성 메시지는 단순 정보 제공이나 제도 안내, 고객 문의에 대한 응답 등 광고·권유 성격이 없는 메시지를 의미합니다.
</정보성 메시지 정의>
"""
    
    agent = Agent()
    result = agent(prompt)
    
    return result.message['content'][0]['text'] 
    
# is_adv 값에 따른 분기 처리
def analyze_messages(df):
    """
    is_adv 값에 따라 메시지를 분석하고 처리하는 함수
    """
    results = []
    
    for idx, row in df.iterrows():
        pk = row['PK']
        title = row['message_title']
        content = row['message_contents']
        is_adv = row['is_adv']
        send_count = row['send_count']
        
        if is_adv == 'Y':
            # 광고성 문자인 경우 - 광고성 사유 분석
            reason = analyze_advertisement_reason(title, content)
            result = {
                'PK': pk,
                'type': '광고성',
                'title': title,
                'content': content[:100] + '...' if len(content) > 100 else content,
                'analysis': f"광고성 문자 사유: {reason}",
                'send_count': send_count
            }
        else:
            # 정보성 문자인 경우 - 개선 제안
            suggestion = suggest_informational_improvement(title, content)
            result = {
                'PK': pk,
                'type': '정보성',
                'title': title,
                'content': content[:100] + '...' if len(content) > 100 else content,
                'analysis': f"개선 제안: {suggestion}",
                'send_count': send_count
            }
        
        results.append(result)
    
    return results
```

```python
print("=== 메시지 분석 시작 ===")
analysis_results = analyze_messages(df)

# DataFrame으로 결과 저장
results_df = pd.DataFrame(analysis_results)
results_df.head()  # 결과 DataFrame의 첫 5행을 출력

```

```python
results_df.to_csv('message_analysis_results.csv', index=False, encoding='utf-8-sig')
print(f"\n분석 결과가 'message_analysis_results.csv' 파일로 저장되었습니다.")
```
